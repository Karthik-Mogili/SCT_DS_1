```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(
    "API_SP.POP.TOTL_DS2_en_csv_v2_40826.csv",
    skiprows=4,
    encoding="latin1"
)

year = "2022"   # <-- manually set year

df = df[["Country Name", year]].dropna()
df_sorted = df.sort_values(by=year, ascending=False).head(10)

df_sorted[year] = df_sorted[year] / 1_000_000

plt.figure(figsize=(10,6))
plt.bar(df_sorted["Country Name"], df_sorted[year])

plt.xticks(rotation=45)
plt.xlabel("Country")
plt.ylabel("Population (in Millions)")
plt.title(f"Top 10 Countries by Population ({year})")

plt.tight_layout()
plt.show()
```
